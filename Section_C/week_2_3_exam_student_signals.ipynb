# Week 2–3 Student Performance Analytics

This notebook analyzes student performance using attendance, diary entries, completed tasks, assignment scores, and lab scores.

```python
import pandas as pd

data = {
    "student_id": ["2026CS001","2026CS002","2026IT001","2026IT002","2026CE001","2026CE002","2026CS003","2026IT003"],
    "name": ["Aditi","Bhavana","Chaitra","Divya","Esha","Fatima","Gauri","Harshada"],
    "department": ["Computer Engineering","Computer Engineering","Information Technology","Information Technology","Civil Engineering","Civil Engineering","Computer Engineering","Information Technology"],
    "attendance_pct": [92,64,78,55,88,71,96,43],
    "diary_entries": [12,6,9,4,10,8,14,2],
    "tasks_completed": [8,5,7,3,8,6,9,2],
    "assignment_score": [84,58,72,46,79,66,91,35],
    "lab_score": [88,61,76,52,82,69,94,41]
}

df = pd.DataFrame(data)

df.to_csv("week_2_3_exam_student_signals.csv", index=False)

print("CSV created successfully!")

df.head()
```

## Dataset Preview

The dataset has been created and the first five rows are displayed.

```python
print("Shape of Dataset:")
print(df.shape)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())
```

## Data Validation

This section checks the dataset shape, data types, and missing values before analysis.

```python
def readiness_score(attendance, diary, tasks, assignment, lab):

    score = (
        attendance * 0.25 +
        diary * 1.5 +
        tasks * 2 +
        assignment * 0.25 +
        lab * 0.25
    )

    if score > 100:
        score = 100
    elif score < 0:
        score = 0

    return round(score, 2)
```

## Readiness Score Function

The readiness score is calculated using attendance, diary entries, tasks completed, assignment score, and lab score.

```python
df["Readiness_Score"] = df.apply(
    lambda row: readiness_score(
        row["attendance_pct"],
        row["diary_entries"],
        row["tasks_completed"],
        row["assignment_score"],
        row["lab_score"]
    ),
    axis=1
)

df
```

## Readiness Score Calculation

The readiness score has been calculated for every student.

```python
performance = []

for score in df["Readiness_Score"]:

    if score >= 85:
        performance.append("Excellent")

    elif score >= 70:
        performance.append("Strong")

    elif score >= 50:
        performance.append("Developing")

    else:
        performance.append("Support")

df["Performance_Band"] = performance

df
```

## Student Classification

Students are classified into Support, Developing, Strong, or Excellent based on their readiness score.

```python
print("Department Average Readiness Score")

department_avg = df.groupby("department")["Readiness_Score"].mean()

print(department_avg)

print("\nTop Performer")

top = df.loc[df["Readiness_Score"].idxmax()]

print(top)

print("\nStudents Needing Support")

support = df[df["Performance_Band"] == "Support"]

print(support)
```
